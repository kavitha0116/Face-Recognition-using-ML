# Open cv

```python
import cv2
import os

# Create a subfolder to store the images
if not os.path.exists('images'):
    os.makedirs('images')

# Initialize the camera
cap = cv2.VideoCapture(0)

# Set camera properties
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# Define the number of images to capture
num_images = 100

# Start capturing images
for i in range(num_images):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # Save the frame to a file
    filename = f"images/image{i+1}.jpg"
    cv2.imwrite(filename, frame)
    
# Release the camera and close all windows
cap.release()
#cv2.destroyAllWindows()
```

```python
import os
from skimage import io, color, transform

# Define the input and output directories
input_dir = 'C:/Users/kavitha/images'
output_dir = 'C:/Users/kavitha/resize_opencv'

# Define the desired image size
new_size = (25, 25)

# Loop through each subfolder in the input directory
for subdir in os.listdir(input_dir):
    # Create a new subdirectory in the output directory with the same name
    subdir_path = os.path.join(output_dir, subdir)
    os.makedirs(subdir_path, exist_ok=True)

    # Loop through each image file in the current subfolder
    for filename in os.listdir(os.path.join(input_dir, subdir)):
        # Read in the image and convert it to grayscale
        img = io.imread(os.path.join(input_dir, subdir, filename))
        img_gray = color.rgb2gray(img)

        # Resize the grayscale image to the desired size
        img_resized = transform.resize(img_gray, new_size)

        # Save the resized image to the output directory with the same filename
        io.imsave(os.path.join(subdir_path, filename), img_resized)
```

# remove background

```python
import os
from rembg import remove

# Define the input and output directories
input_dir = "C:/Users/kavitha/resize_opencv"
output_dir = "C:/Users/kavitha/rembg_opencv"

# Recursively loop through all subfolders and files in the input directory
for root, dirs, files in os.walk(input_dir):
    for file in files:
        # Check if the file is an image
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            # Define the input and output file paths
            input_path = os.path.join(root, file)
            output_path = input_path.replace(input_dir, output_dir)

            # Create the output directory if it doesn't exist
            os.makedirs(os.path.dirname(output_path), exist_ok=True)

            # Load the input image and remove the background
            with open(input_path, "rb") as f:
                input_image = f.read()
            output_image = remove(input_image)

            # Save the output image
            with open(output_path, "wb") as f:
                f.write(output_image)
```

# generate csv

```python
import os
import csv
import numpy as np
from PIL import Image

# Define the path to the directory containing the subfolders of images
directory_path = 'C:/Users/kavitha/rembg_opencv'

# Define the name of the output CSV file
output_csv_name = 'C:/Users/kavitha/open_cv.csv'

# Define the full path to the output CSV file
output_csv_path = os.path.join(directory_path, output_csv_name)

# Loop through the subfolders and their contents, and write the flattened image data to the CSV file
with open(output_csv_path, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Class'] + list(range(25*25))) # Header row
    for class_name in os.listdir(directory_path):
        class_path = os.path.join(directory_path, class_name)
        if not os.path.isdir(class_path):
            continue
        for filename in os.listdir(class_path):
            file_path = os.path.join(class_path, filename)
            try:
                img = Image.open(file_path)
                img = img.resize((25, 25)) # Resize the image to 28x28
                img_array = np.array(img.convert('L'), dtype=np.uint8) # Convert to grayscale and convert to NumPy array
                img_flat = img_array.flatten() # Flatten the array
                
                writer.writerow([class_name] + list(img_flat))
            except:
                print('Error processing image:', file_path)

print('Image data saved to', output_csv_path)
```
