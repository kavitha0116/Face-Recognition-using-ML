```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the training and test data
train_data = pd.read_csv('C:/Users/kavitha/open_cv.csv')
test_data = pd.read_csv("C:/Users/kavitha/test.csv")

# Split the training data into features (X) and target (y)
X = train_data.drop(columns=["Class"]).values
y = train_data["Class"].values

# Define the number of folds and the classifier
n_splits = 10
clf = KNeighborsClassifier(n_neighbors=7)

# Initialize the stratified k-fold splitter
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize the arrays to store the accuracies and indices
split_acc = []
best_acc = 0
best_train_indices = []
best_test_indices = []

# Loop over the folds
for fold, (train_indices, test_indices) in enumerate(skf.split(X, y)):
    # Split the data
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]

    # Train the classifier
    clf.fit(X_train, y_train)

    # Evaluate the classifier on the test set
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    # Store the results
    split_acc.append(acc)
    if acc > best_acc:
        best_acc = acc
        best_train_indices = train_indices
        best_test_indices = test_indices

    # Print the results for this fold
    print(f"Fold {fold+1}, accuracy = {acc:.3f}")

# Print the average accuracy over all folds
print(f"Average accuracy = {sum(split_acc)/n_splits:.3f}")

# Train the classifier on the best training indices
X_train_best = X[best_train_indices]
y_train_best = y[best_train_indices]
clf.fit(X_train_best, y_train_best)

# Evaluate the classifier on the test set
X_test = test_data.drop(columns=["Class"]).values
y_test = test_data["Class"].values
y_pred_test = clf.predict(X_test)
test_acc = accuracy_score(y_test, y_pred_test)
print(f'Training indices: {best_train_indices}')
print(f'Testing indices: {best_test_indices}')
print(f"Test accuracy = {test_acc:.3f}")




# Confusion matrix for training indices
train_predictions = clf.predict(X_train_best)
conf_matrix = confusion_matrix(y_train_best, train_predictions)

n_classes = conf_matrix.shape[0]
specificity = np.zeros(n_classes)
sensitivity = np.zeros(n_classes)
precision = np.zeros(n_classes)
f_score = np.zeros(n_classes)
balanced_accuracy = np.zeros(n_classes)

for i in range(n_classes):
    print("Confusion Matrix:\n", conf_matrix)
    true_positives = conf_matrix[i, i]
    false_positives = np.sum(conf_matrix[:, i]) - true_positives
    false_negatives = np.sum(conf_matrix[i, :]) - true_positives
    true_negatives = np.sum(conf_matrix) - true_positives - false_positives - false_negatives
    specificity[i] = true_negatives / (true_negatives + false_positives)
    sensitivity[i] = true_positives / (true_positives + false_negatives)
    precision[i] = true_positives / (true_positives + false_positives)
    f_score[i] = (2 * precision[i] * sensitivity[i]) / (precision[i] + sensitivity[i])
    balanced_accuracy[i] = (specificity[i] + sensitivity[i]) / 2
    print(f'Class {i+1}: Specificity={specificity[i]:.2f}, Sensitivity={sensitivity[i]:.2f}, Precision={precision[i]:.2f}, F-score={f_score[i]:.2f}, Balanced Accuracy={balanced_accuracy[i]:.2f}')

    fpr, tpr, thresholds = roc_curve(y_train_best, train_predictions, pos_label=1)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')

    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.show()

# Confusion matrix for testing indices
conf_matrix2 = confusion_matrix(y_test, y_pred_test)
n_classes = conf_matrix2.shape[0]
specificity = np.zeros(n_classes)
sensitivity = np.zeros(n_classes)
precision = np.zeros(n_classes)
f_score = np.zeros(n_classes)
balanced_accuracy = np.zeros(n_classes)

for i in range(n_classes):
    print("Confusion Matrix:\n", conf_matrix2)
    true_positives = conf_matrix2[i, i]
    false_positives = np.sum(conf_matrix2[:, i]) - true_positives
    false_negatives = np.sum(conf_matrix2[i, :]) - true_positives
    true_negatives = np.sum(conf_matrix2) - true_positives - false_positives - false_negatives
    specificity[i] = true_negatives / (true_negatives + false_positives)
    sensitivity[i] = true_positives / (true_positives + false_negatives)
    precision[i] = true_positives / (true_positives + false_positives)
    f_score[i] = (2 * precision[i] * sensitivity[i]) / (precision[i] + sensitivity[i])
    balanced_accuracy[i] = (specificity[i] + sensitivity[i]) / 2
    print(f'Class {i+1}: Specificity={specificity[i]:.2f}, Sensitivity={sensitivity[i]:.2f}, Precision={precision[i]:.2f}, F-score={f_score[i]:.2f}, Balanced Accuracy={balanced_accuracy[i]:.2f}')

    fpr, tpr, thresholds = roc_curve(y_test, y_pred_test, pos_label=1)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')

    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.show()

```
